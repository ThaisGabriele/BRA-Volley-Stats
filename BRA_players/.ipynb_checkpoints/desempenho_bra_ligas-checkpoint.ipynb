```python
def organiza_dfs(df):
    df = df.rename(columns={'Unnamed: 0_level_1':'0', 'Unnamed: 1_level_1':'1'},level=1)
    df = df.drop(labels=[0,1], axis=0)
    df.columns = df.columns.map('_'.join)
    df = df.set_index('Round_0')
    df['Ataque_Tot'] = df['Ataque_Tot'].astype(float)
    df['Ataque_Exc.'] = df['Ataque_Exc.'].astype(float)
    att = lambda x : round((df['Ataque_Exc.']  / df['Ataque_Tot']) * 100,2)
    df['Ataque_% Exc.'] = att('Ataque_% Exc.')
    df = df.fillna(0)
    df['Ataque_Tot'] = df['Ataque_Tot'].astype(int)
    df['Ataque_Exc.'] = df['Ataque_Exc.'].astype(int)
    return df
```

```python
df_nia = organiza_dfs(df_nia)
```

```python
df_lorrayna = organiza_dfs(df_lorrayna)
```

```python
df_martinez = organiza_dfs(df_martinez)
```

```python
plt.figure(figsize=(8,2))
ax = sns.boxplot(x=df_nia['Ataque_% Exc.'])
```

```python
plt.figure(figsize=(8,2))
ax = sns.boxplot(x=df_lorrayna['Ataque_% Exc.'])
```

## Liga Turca

```python
## progressão Gabi 
df_gabi = pd.read_html("https://tvf-web.dataproject.com/Statistics_AllPlayers.aspx?ID=62&PID=114&Player=3979", encoding='utf-8')
```

```python
df_gabi = df_gabi[12]
```

```python
df_gabi.columns
```

```python
# renomeando colunas - TURCO -> PT
df_gabi = df_gabi.rename(columns={'Etap':'Round',
                'Oynanmış set': 'Setsjogados',
                'Sayılar': 'Pontos',
                'Servis':'Serviço',
                'Servis karşılama':'Recepção',
                'Hücum':'Ataque',
                'BLK':'Bloqueio'
                
},level=0)
```

```python
df_gabi = df_gabi.rename(columns={'Unnamed: 0_level_1':'0',
                                  'Unnamed: 1_level_1':'1',
                                  'K':'V', 
                                  'SK':'BP', 
                                  'Kaz.-Kay.':'V-P',
                                  'Hata':'Err',
                                  'Set başına ace':'Ace por set', 
                                  'Ver':'Efi',
                                  'Mük.':'Exc.', 
                                  'Mük. %': 'Exc.%', 
                                  'Mük.%':'Exc.%',
                                  'Blok':'Blk',
                                  'File':'Rede',
                                  'Say':'Pts',
                                  'Set başına sayı':'Ponto por Set'},level=1)
```

```python
df_gabi
```

```python
df_gabi.columns
```

```python
def organiza_df(df):
    df.columns = df.columns.map('_'.join)
    df = df.drop(labels=[0,1], axis=0)
    df = df.set_index('Round_0')
    df['Ataque_Tot.'] = df['Ataque_Tot.'].astype(float)
    df['Ataque_Exc.'] = df['Ataque_Exc.'].astype(float)
    df['Ataque_Efi'] = df['Ataque_Efi'].astype(float)
    att = lambda x : round((df['Ataque_Exc.']  / df['Ataque_Tot.']) * 100,2)
    df['Ataque_Exc.%'] = att('Ataque_Exc.%')
    df = df.fillna(0)
    df['Ataque_Tot.'] = df['Ataque_Tot.'].astype(int)
    df['Ataque_Exc.'] = df['Ataque_Exc.'].astype(int)
    
    return df
```

```python
df_gabi = organiza_df(df_gabi)
```

```python
df_gabi['Ataque_Exc.%']
```

```python
df_ataque = df_gabi.filter(items=['Ataque_Tot.','Ataque_Exc.','Ataque_Exc.%','Ataque_Efi'])
```

```python
df_ataque
```

```python
plt.figure(figsize=(8,4))

ax = sns.boxplot(x=df_ataque['Ataque_Exc.%'])

plt.show()
```
